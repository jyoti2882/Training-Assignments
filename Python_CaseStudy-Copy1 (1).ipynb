```python
from datetime import date
class Book_Ratings:
    def __init__(self,rating=[]):
        self.ratings= []
        self.serial_number=0 
    
    def add_rating(self,user_id, rating):
        self.ratings.append((user_id,rating))
            

class Books(Book_Ratings):
    def __init__(self, isbn, title, author, year_of_publication, publisher, image_url_s, image_url_m, image_url_l):
        self.isbn= isbn
        self.title= title
        self.author= author
        self.year_of_publication= year_of_publication
        self.publisher= publisher
        self.image_url_s= image_url_s
        self.image_url_m= image_url_m
        self.image_url_l= image_url_l
        self.issued_to= None
        self.issue_date= None
        super().__init__()
           

class Users:
    def __init__(self,serial_number,address,area_code):
        self.serial_number= serial_number
        self.address= address
        self.area_code= area_code
    
    
class Library:
    def __init__(self):
        self.books_details={}
        self.users_details= {}
        
    
    def add_book(self,isbn, title, author, year_of_publication, publisher, image_url_s, image_url_m, image_url_l):
        b= Books(isbn, title, author, year_of_publication, publisher, image_url_s, image_url_m, image_url_l)
        self.books_details[isbn]= b
        print("A new book is added to the Collection")
        
        
    def delete_book(self,isbn):
        if len(self.books_details) ==0 :
            print("No Book is present in Library")
            return 
        book_deleted= self.books_details.pop(isbn)
        print("{} is removed from the collection".format(book_deleted.title))
        
        
    def update_book(self,isbn,new_title):
        if isbn not in self.books_details.keys():
            print("Book with the given ISBN is not present")
        else:
            self.books_details[isbn].title= new_title
            print("Title of the Book is updated")
        
    def check_title(self,title):
        flag=0
        for v in self.books_details.values():
            if title == v.title:
                print("Book is present")
                flag=1
                break
        if flag==0:    
            print("Book is not present")
                
            
    def check_isbn(self,isbn):
        if isbn in self.books_details.keys():
            print("ISBN is present")
        else:
            print("ISBN not present")
            
    def add_user(self,serial_number,address,area_code):
        u= Users(serial_number,address,area_code)
        self.users_details[serial_number]= u
        print("User is added")
        
        
    def change_user(self,serial_number,updated_value):
        if len(updated_value) == 2:
            address= updated_value[0]
            area_code= updated_value[1]
            self.users_details[serial_number].address= address
            self.users_details[serial_number].area_code= area_code
            print("Updated the record")
        elif len(updated_value) == 1:
            if isinstance(updated_value[0],str):
                self.users_details[serial_number].address= updated_value[0]
            else:
                self.users_details[serial_number].area_code= updated_value[0]
            print("Updated the record")
        else:
            print("No values sent for update")
            
            
    def issue(self, isbn, serial_number, issue_date):
        if len(self.books_details) ==0:
            print("No Book is present in Library")
            return
        if isbn in self.books_details and serial_number in self.users_details:
            if self.books_details[isbn].issued_to == None:
                self.books_details[isbn].issued_to= serial_number
                self.books_details[isbn].issue_date= issue_date
                print("Book Issued Successfully")
            else:
                print("The provided Book is currently not available")
                   
        else:
            print("Either the book is not present or user is not valid")
                
                
    def return_book(self, isbn, date_of_return):
        if self.books_details[isbn].issue_date == None:
            print("Book was not issued")
            return
        
        self.books_details[isbn].issued_to = None
        d1= self.books_details[isbn].issue_date
        days= (date_of_return - d1).days
        late_fees=0
        if days>7:
            late_fees= (days-7)*10
        print("Book returned Succesfully")
        if late_fees>0:
            print("Your Fine for late return is: ",late_fees)
        

    
    def avg_rating(self):
        if len(self.books_details)==0:
            print("No Books Present in the Catalog")
            return
        temp= dict()
        for v in self.books_details.values():
            if v.publisher in temp:
                
                try:
                    l=len(v.ratings)
                    s=0
                    for t in v.ratings:
                        s+=t[1]
                    temp[v.publisher].append(s/l)
                except ZeroDivisionError:
                    temp[v.publisher].append(0)
            else:
                try:
                    l=len(v.ratings)
                    s=0
                    for t in v.ratings:
                        s+=t[1]
                    temp[v.publisher]= [s/l]
                except ZeroDivisionError:
                    temp[v.publisher]= [0]
                    
        for k,v in temp.items():
            print("Ratings for publisher {} is {}".format(k,sum(v)/len(v)))
            
            
    def get_Total(self):
        if len(self.books_details)==0:
            print("No Books Present in the Catalog")
            return
        temp= dict()
        for v in self.books_details.values():
            if v.year_of_publication in temp:
                temp[v.year_of_publication]+=1
            else:
                temp[v.year_of_publication]=1
        
        for k,v in temp.items():
            print("Total number of books published in year {} is {}".format(k,v))
        
    
    def get_Top_Five(self):
        if len(self.books_details)==0:
            print("No Books Present in the Catalog")
            return
        temp= dict()
        for v in self.books_details.values():
            s=0
            for t in v.ratings:
                s+=t[1]
            temp[v.isbn]= s
        result= sorted(temp.items(), key=lambda x:x[1], reverse=True)[0:5]
        print("Our Top five books are:")
        for t in result:
            print(t[0],self.books_details[t[0]].title)
            
    def view_Catalog(self):
        if len(self.books_details)==0:
            print("No Books Present in the Catalog")
        else:
            print("Books present in our Catalog:")
            for k,v in self.books_details.items():
                print("ISBN: {}  BookName: {}  Publisher: {}  Year Of Publication: {}".format(k, v.title, v.publisher, v.year_of_publication ))
    
    def display(self):
        if len(self.books_details) == 0:
            print("No Books Present in the Catalog")
        else:
            print("Books present in our Catalog with Ratings:")
            for k, v in self.books_details.items():
                print("ISBN: {} BookName: {}".format(k, v.title))
                if len(v.ratings) > 0:
                    print("Ratings:")
                    for user_id, rating in v.ratings:
                        print("USer Id: {}, Area Code : {}, Rating: {}".format(user_id, self.users_details[user_id].area_code, rating))
                else:
                    print("No ratings available for this book")
    
        
            

if __name__ == "__main__":
    l = Library()
    
    while True:
        print("\nLibrary Management System Menu:")
        print("1. Add Book")
        print("2. Delete Book")
        print("3. Update Book")
        print("4. Check Title")
        print("5. Check ISBN")
        print("6. Add User")
        print("7. Change User")
        print("8. Issue Book")
        print("9. Return Book")
        print("10. Add Rating to a book")
        print("11. Average Rating")
        print("12. Total Books Published per Year")
        print("13. Top Five Books")
        print("14. View Catalog")
        print("15. Display")
        print("0. Exit")
        
        choice = int(input("Enter your choice: "))
        
        if choice == 0:
            print("Exiting Library Management System.")
            break
            
        elif choice == 1:
            isbn= input("Enter isbn: ")
            title= input("Enter title of Book: ")
            author= input("Enter name of author: ")
            year_of_publication= int(input("Enter year of publication of book: "))
            publisher= input("Enter name of publisher: ")
            image_url_s= input("Enter small image url: ")
            image_url_m= input("Enter medium image url: ")
            image_url_l= input("Enter large image url: ")
            l.add_book(isbn, title, author, year_of_publication, publisher, image_url_s, image_url_m, image_url_l)
            
        elif choice == 2:
            isbn= input("Enter isbn of Book to be Deleted: ")
            l.delete_book(isbn)
            
        elif choice == 3:
            isbn= input("Enter isbn of Book to be updated: ")
            new_title= input("Enter the new title for book: ")
            l.update_book(isbn, new_title)
        
        elif choice == 4:
            title= input("Enter the title for book you want to check is present or not: ")
            l.check_title(title)
            
        elif choice ==5:
            isbn= input("Enter the isbn for book you want to check is present or not: ")
            l.check_isbn(isbn)
            
        elif choice ==6:
            serial_number= int(input("Enter Serial Number: "))
            address= input("Enter user address: ")
            area_code= int(input("Enter Area Code: "))
            l.add_user(serial_number,address,area_code)
            
        elif choice == 7:
            user_id= int(input("Enter serial number of user whoose details you want to update:  "))
            print("What do you want to update:")
            print("1. Address")
            print("2. Area Code")
            print("3. Both")
            ch= int(input("Enter Your Choice: "))
            if ch ==1:
                add= input("Enter new address: ")
                l.change_user(user_id,[add])
            if ch ==2:
                ac= int(input("Enter new area code: "))
                l.change_user(user_id,[ac])
            if ch ==3:
                add= input("Enter new address: ")
                ac= int(input("Enter new area code: "))
                l.change_user(user_id,[add,ac])
    
            
        elif choice ==8:
            l.view_Catalog()
            isbn= input("Enter ISBN of Book you want to Issue: ")
            user_id= int(input("Enter serial number of user:  "))
            l.issue(isbn, user_id, date.today())
            
        elif choice == 9:
            isbn= input("Enter ISBN of Book you want to Return: ")
            date_of_return= date.today()
            l.return_book(isbn, date_of_return)
            
        elif choice == 10:
            isbn= input("Enter ISBN of Book you want to give rating: ")
            user_id= int(input("Enter Users Serial number"))
            rating= int(input("Enter rating between 1-10: "))
            for k,v in l.books_details.items():
                if isbn == v.isbn:
                    v.add_rating(user_id,rating) 
                    
        elif choice == 11:
            l.avg_rating()
            
        elif choice == 12:
            l.get_Total()
        
        elif choice == 13:
            l.get_Top_Five()
            
        elif choice == 14:
            l.view_Catalog()
            
        elif choice == 15:
            l.display()
            
        else:
            print("Invalid Choice")
            
```
